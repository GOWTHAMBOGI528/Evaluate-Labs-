```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Age'], kde=True, bins=30, color='teal')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(data=df, x='Survived', palette='coolwarm')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='mako')
plt.title("Passenger Class vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Pclass', y='Fare', palette='viridis')
plt.title("Boxplot: Fare by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='deep')
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

```python
print("\n----- Observations -----")
print("1️⃣ Female passengers and 1st class travelers had higher survival rates.")
print("2️⃣ Younger passengers and families with fewer members had better chances of survival.")
print("3️⃣ There is a moderate positive correlation between Fare and Survival.")
print("4️⃣ Pclass and Survival are negatively correlated (lower class → lower survival).")
print("5️⃣ Median passenger age is around 28 years.")

# ================== Summary ==================
print("\n----- Summary -----")
print("The EDA reveals clear patterns in passenger demographics and survival rates. "
      "Women and upper-class passengers were more likely to survive, "
      "while lower-class males had lower survival chances. "
      "Fare and Class play key roles in determining survival probability.")
```
